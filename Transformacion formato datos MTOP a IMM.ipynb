```python
### Arreglar

def to_IMM_format(cols,path= r'c:\Python\Master Datascience\Tesis\dataset\Tránsito PP91 R0 5K000\2017',remove_duplicate=False,group=False):
    """Open data csvs and returns specific dataframes"""
    all_files = glob.glob(path + "/*.csv")
    li = []
    print('Total archivos',len(all_files))
    counter = 1
    for filename in all_files:
        print('Van ',counter,' archivos')
        counter +=1
        try:
            if remove_duplicate: # Removes duplicate values if specificated
                df.drop_duplicates(subset=cols,keep='first',inplace=True)
            df = pd.read_csv(filename, index_col=None, header=0, encoding="ISO-8859-1",usecols=cols,sep=";")
            df['mix'] = df.Fecha + ' ' + df.Hora
            df.mix = pd.to_datetime(df.mix,format='%d/%m/%Y %H:%M:%S.%f')
            df.drop(['Fecha','Hora'],axis=1,inplace=True)
            df.rename(columns={"mix": 'Fecha_Hora'},inplace=True)
            df.set_index('Fecha_Hora',inplace=True)
            if group:
                df_tm = df.groupby(pd.Grouper(freq='5Min')).count()
                li.append(df_tm)
            else:
                li.append(df)
        except Exception as e:
            print(e)
    return pd.concat(li, axis=0)
```

```python
def adapt_to_database(df):
    """Adapts MTOP dataframe to be store in sqlite3 database with IMM data"""
    df.Fecha_Hora = pd.to_datetime(df.Fecha_Hora,format='%Y/%m/%d %H:%M:%S')
    df.sort_values(by='Fecha_Hora',inplace=True)
    df.set_index('Fecha_Hora',inplace=True)
    df['volumen_hora'] = df.rolling('H').sum() 
    ### Add default data to dataframe
    df['latitud'] = -34.864379
    df['longitud'] =  -56.252485
    df['id_camara'] = 999
    df['id_carril'] = 999
    df['dsc_avenida'] = 'RUTA NAL INTERCONEXION RUTA 1 Y 5'
    df['dsc_int_anterior'] = 'AV DR SANTIN CARLOS ROSSI'
    df['dsc_int_siguiente'] = 'BV GRAL ARTIGAS'
    df['id_corregido'] = 276
    df.rename(columns={'Numero del carril':'volumen'},inplace=True)
    df.reset_index(inplace=True)
    print(df.columns)
    return df
```

```python
def create_database():
    """Adds all MTOP dataframes to sqlite3 database with IMM Data"""
    #engine = create_engine('sqlite://', echo=False)
    all_files = (['MTOP_2017_formato_IMM.csv','MTOP_2018_formato_IMM.csv',
                  'MTOP_2019_formato_IMM.csv','MTOP_2020_formato_IMM.csv'])
    con = sqlite3.connect('conteo_vehicular_id_corregido.sqlite')
    counter = 0
    for filename in all_files:
        print("Van {}, faltan {}".format(counter,len(all_files)-counter))
        df = pd.read_csv(filename, index_col=None, header=0, encoding="ISO-8859-1")
        df = adapt_to_database(df)
        df.to_sql('data', con=con, if_exists='append',index=False)
        counter += 1
    con.commit()
create_database()
print('Exito')
```
